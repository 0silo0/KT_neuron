```python
plt.figure(figsize=(10, 6))
sns.histplot(diabetes_012['BMI'], kde=True, color='blue')
plt.title('Распределение индекса массы тела (BMI)')
plt.xlabel('BMI')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(diabetes_012['BMI'], color='green')
plt.title('Boxplot индекса массы тела (BMI)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='PhysActivity', y='BMI', hue='Diabetes_012', data=diabetes_012)
plt.title('Влияние физической активности на BMI у людей с разными стадиями диабета')
plt.xlabel('Физическая активность (0 = нет, 1 = да)')
plt.ylabel('Индекс массы тела (BMI)')
plt.legend(title='Диабет')
plt.show()
```

```python
diab_chek1 = diabetes_012
diab_chek1['Smoker'] = diab_chek1['Smoker'].astype(int)
diab_chek1['HighBP'] = diab_chek1['HighBP'].astype(int)

plt.figure(figsize=(12, 6))
sns.countplot(x='Smoker', hue='HighBP', data=diab_chek1, palette='coolwarm')
plt.title('Влияние курения на высокое давление среди людей с разными стадиями диабета')
plt.xlabel('Курение (0 = нет, 1 = да)')
plt.ylabel('Количество людей')
plt.legend(title='Высокое давление (HighBP)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='PhysHlth', hue='Diabetes_012', data=diabetes_012, palette='Set1')
plt.title('Влияние возраста на физическое здоровье у людей с разными стадиями диабета')
plt.xlabel('Возрастная группа')
plt.ylabel('Физическое здоровье (количество дней с проблемами)')
plt.legend(title='Диабет')
plt.show()
```
